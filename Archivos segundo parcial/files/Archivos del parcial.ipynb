```python
"""
Posibles pacientes con enfermedades cardiacas 
"""

archivoheart = pd.read_csv("heart.csv")
archivoheart
```

```python
#ejercicio 2.1

edades=archivoheart[["edad","blanco"]]
edades

filtroblancosi = archivoheart["blanco"] == 1
filtroblancono = archivoheart["blanco"] == 0
blancospositivos=archivoheart[filtroblancosi]
blancosnegativos=archivoheart[filtroblancono]
mediaedadposi=blancospositivos['edad'].mean()
desvioedadposi=blancospositivos['edad'].std()
mediaedadnega=blancosnegativos['edad'].mean()
desvioedadnega=blancosnegativos['edad'].std()

print("el promedio de edad para personas candidatas a tener alguna patologia cardiaca es", mediaedadposi, "el desvio estandar es",desvioedadposi)
print("el promedio de edad para personas que no son candidatas a tener alguna patologia cardiaca es", mediaedadnega, "el desvio estandar es",desvioedadnega)
```

```python
colesterolenposi= blancospositivos["nivelCol"]
colesterolnegati= blancosnegativos["nivelCol"]
valmaxcolposi= colesterolenposi.max()
valmincopoi=colesterolenposi.min()
valmaxcolnega= colesterolnegati.max()
valmincolnega= colesterolnegati.min()

print("Valor maximo de colesterol en personas candidatas a patologias cardiacas es ",  valmaxcolposi)
print("Valor minimo de colesterol en personas candidatas a patologias cardiacas es ",  valmincopoi)
print("Valor maximo de colesterol en personas no candidatas a patologias cardiacas es ",  valmaxcolnega)
print("Valor minimo de colesterol en personas  no candidatas a patologias cardiacas es ",  valmincolnega)
```

```python
filtroglucemia = archivoheart["glucA"] == 1
persprediab= archivoheart[filtroglucemia]
cantper= len(persprediab)

print("cantidad de personas prediabeticas son", cantper)
```

```python
import matplotlib.pyplot as plt
# Parte 3
plt.hist(blancospositivos['edad'], bins=80, color="blue")
plt.hist(blancosnegativos['edad'], bins=80, color="red")
plt.title("Histograma de edades")
plt.ylabel("cantidad de personas")
plt.ylim((0,30))
plt.xlabel("edades")
plt.legend()
plt.show()
```

```python
plt.hist(blancospositivos['presRest'], bins=80, color="blue")
plt.hist(blancosnegativos['presRest'], bins=80, color="red")
plt.title("Histograma de presion")
plt.ylabel("cantidad de presion")
plt.ylim((0,30))
plt.xlabel("presion")
plt.legend()
plt.show()
```

```python
#blancospositivos['maxHR']

fig, ax = plt.subplots()

box = ax.boxplot([blancospositivos['maxHR'],blancosnegativos['maxHR']], widths = 0.5, patch_artist = True, showbox = True, showfliers = True,
          showmeans = False)
ax.set_xticklabels(["Frecuencia de positivos", "Frecuencia de negativos"])
ax.set_ylabel("frecuencia")
ax.set_title("Boxplots para positovos y negativos")

```
